### 1. 개발자에게 영어가 중요한 이유

> 개발자에게 개발언어만큼 중요한것은 영어다.
- IT 교육기관에서 중요하게 교육해 주길 바라는 내용 :<br> 영어 독해력이 높은 순위를 차지

> 지식과 정보의 보고
- 프로그래밍 언어가 영어권에서 개발
- 의사소통을 위한 작명 관례가 중요
- 깨끗한 코드의 중요성

> 의사소통의 도구
- 공동체 참여
- 묻고 답하기(Stack Overflow, Quora), 메일링 리스트 

> 주의사항
- 문해력의 중요성
- 영어를 제대로 학습하고 제대로 사용해야 한다 !


### 2. 영어 학습 과정에서 무엇에 집중해야 할까?
> 읽기가 중요한 이유
- 4대 능력 : 말하기, 듣기, 쓰기, 읽기
- 소프트웨어 개발에서 대다수 시간은 생각과 읽기에 사용

> 읽기 좋은 코드가 중요한 이유
- 코드는 이해하기 쉬워야 한다
- 코드 이해도가 핵심
    - 다른 사람이 이해하는 시간을 줄여야함
    - 구현과 테스트가 쉬운 코드의 특성
    - 가독성 높은 코드는 클린코드의 기본 특성
- 코드 자체가 문서로 볼 수 있다!
    - 코드 자체를 문서로 생각해야
    - 생산성을 높이는 첫 걸음

> 코드 작성과정에서 쓰기의 중요성
- 코드 내부 식별자는 영어가 일반적이다
    - 타입, 변수, 메서드, 함수, 모듈, 라이브러리, 네임스페이스
- 코드 내부 주석도 영어가 일반적
    - 문서화를 위한 주석
- 코드 내부 오류 메시지도 영어가 일반적이다
- 형상 관리를 위한 코드 커밋 메시지도 영어가 일반적이다
    - 코드 변경의 이유와 배경, 그리고 부수 효과 등을 설명하는 문구
        - how 는 커밋에 쓰지 않는다. how는 코드에 있다!

> 코드와 글의 공통점
- 그레디 부치(객체지향 대가)
    - 깨끗한 코드는 단순하고 직접적이다.
    - 깨끗한 코드는 잘 쓴 문장처럼 읽힌다.
    - 깨끗한 코드는 결코 설계자의 의도를 숨기지 않는다.
    - 오히려 명쾌환 추상화와 단순한 제어문으로 가득하다.

- 워드 커닝엄(위키 창시자)
    - 코드를 읽으면서 짐작했던 기능을 각 루틴이 그대로 수행한다면 깨끗한 코드라 불러도 됨
    - 코드가 그 문제를 풀기 위한 언어처럼 보인다면 아름다운 코드!

> 그리고 어휘력
- 소프트웨어 개발 과정에서 다양한 문서를 읽어야한다
- 전공이나 직업 관점에서 생각해보자
    - 전문 분야에서 두각을 나타내려면 어휘력이 매우 중
- 단어를 얼마나 알아야할까
    - 모국어 사용자라면 최소 2만 단어는 알고 있다
    - 어휘력 강화에는 왕도도 없고 지름길도 없다

> 간단 정리
- 가장 먼저 읽기에 투자
    - 기술 자료와 코드를 읽을 수 있으면 된다
- 그리고 쓰기에 투자
    - 실무에 적합한 수준을 달성하면 된다
- 어휘력은 그야말로 기본 중의 기본으로 생각해야 한다
    - 특히 기술 용어에 대해 익숙해져야한다
    - 다행히도 기술 자료에 사용되는 영어 단어는 생각보다 어렵지 않다

### 3. 개발자를 위한 영어 학습 로드맵
> 자원의 밀도
- 구글에는 무지막지한 양의 정보가 나온다
- 초심자때는 한국어/영어 자료가 풍부
- 중급자때부터 한국어 자료는 찾기 어렵고, 영어 자료도 희소해지기 시작
    - 이때, 영어가 익숙하지 않으면 매우 힘들다
- 고급자가 되면 영어자료가 늘어나며, 인기 있는 자료는 한국어로도 번역됨

> 학습 역량 향상에 초점을 맞추자
- 단기적으로 구문을 빠르게 분석할 수 있는 문법 강화
    - 제 2외국어라 불리는 상황을 극복하기 위한 노력이 필요
    - 회화나 비즈니스 영어와는 또 다른 접근 방법이 필요
- 중장기적으로 어휘력 강화
    - 기술 용어에 익숙해지기
    - 개발 문서에서 자주 증장하는 어휘와 표현법에 익숙해지기
- 읽기 연습
    - 실무에 필요한 읽기 연습을 진행할 필요가 있음
    - 수준에 맞는 읽을 거리를 찾아서 꾸준하게 시간을 투자해야함!

> 개발 역량 향상으로 초점을 옮겨가자
- 남의 코드를 제대로 읽을 수 있게 준비
    - 실제 프로그램 논리, 주석, 커밋 메시지, 영어로 된 각종 문서
    - 오픈소스를 적극적으로 활용
        - 내가 아는 분야의 오픈소스를 봐라!
        - 완성도가 높은 버전을 보지 말어라. 초창기 버전을 봐 ! 
            - 1) 코드양이 많다 
            - 2) 예외처리 모든 해석이 어렵다
- 내 코드를 깔끔하게 만들 수 있게 준비
    - 클린코드가 가능하려면 프로그램 논리도 명쾌해야하지만 영어 관련 역량도 좋아야함
    - 읽기에서 쓰기로 확장
- 문서화에도 신경을 써야함
    - readme나 주석은 영어로 작성하는 습관을 들여야함
    - changelog 등은 커밋 메시지와 연계에서 제대로 작성해야함

> 이 강의에서 다룰 로드맵
- 독해(일기/어휘)
    - 독해 역량 향상
    - 어휘 암기 팁
    - 기술 뉴스 / 튜토리얼 / 매뉴얼 / 기수 블로그 / 기술 사양서 읽는 방법
- 코드 문서화(쓰기/어휘)
    - 영어 작명법
    - 영어 주석 작명법
    - 영어 커밋 메시지 작성법
    - 영어 오류 메시지 작성법
- 정보 습득(읽기)
    - 검색 엔진 활용법
    - sns 활용법
    - 실전 사례
- 질문과 대답(읽기/쓰기)
    - 좋은 질문과 나쁜 질문
    - 커뮤니티 활용법
    - 실전 사례

> 마무리
- 메타 지식에 매몰되지 말자
    - 영어와 관련된 학습에 대한 연구와 탐험도 중요하지만
    - 실제 영어에 익숙해진 다음에야 결과론적으로 학습 방법이 맞는지 아닌지 알 수 있다
- 결론 : "최대한 많이 읽고 많이 쓰고 생각하기"
